# CROSS VALIDATION 

# Cross  Validation  - bu modelni bitta train test bo`lishiga qarab qo`ymasdan uni bir nechta marta turli bo`linmalarda sinab ko`rish jarayoni

# Nega cross validation kerak? 
- agarda datastni faqat bir marta train.test qilib baholasak
- natija tasodifga bo`gliq bo`lihsi mumkin
- model overfitting yoki underfitting qilganini bilib olishimiz qiyin

Cross validation esa: 
- barqaror baho beradi 
- kichik datasetlarda juda foydali 
- modelni tanlashda adolatli solishtirish imkonini beradi 


# Ishlashi 
- Dataset bir nechta qismlarga (fold) bo`linadi 
Har safar 
- bitta qism test qilinadi 
- qolganlari - train qilinadi 

- natijada o`rtacha baholash qiymati olinadi 

# Cross Validation Turlari 
- K-Fold 
- Stratified K-Fold 
- Leave-one-out (LOO)
- Time Series Split 

```python
import pandas as pd 
import numpy as np 

from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression


df = pd.read_csv(r"C:\Users\Rasulbek907\Desktop\Project_One\Data\Raw_Data\titanic.csv")
```

```python
df.head()
```

```python
df = df[["Survived", "Pclass", "Sex", "Age", "Fare"]]

df["Age"].fillna(df["Age"].median(), inplace=True)
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
```

```python
x = df.drop("Survived", axis=1)
y = df["Survived"]
```

# K Fold 

```python
kfold = KFold(
    n_splits=5,   # K = 5  Tavsiya qilinadi 5 dan 10 gacha bo`lish
    shuffle=True,    # datasetni aralashtirish 
    random_state=42
)
```

```python
pipeline =  Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000))
])
```

# Cross validation yordamida baholash

```python
scores = cross_val_score(
    estimator=pipeline,
    X = x, 
    y = y,
    cv= kfold,
    scoring="accuracy"
)
```

```python
print("Har bir fold accuracy natijasi: ", scores)
print("O`rtacha accuracy: ", scores.mean())
print("Standart og`ish: ", scores.std())
```

# Stratified K-Fold Validation

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)


scores = cross_val_score(
    pipeline,
    x,
    y,
    cv=skf,
    scoring="accuracy"
)

scores.mean()
```
