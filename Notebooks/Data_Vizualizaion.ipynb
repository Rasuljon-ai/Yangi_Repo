```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px 
from pandas.plotting import (
    scatter_matrix, parallel_coordinates, andrews_curves, lag_plot
)

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (6,4)
```

# Real datasetlarni chaqirib olish 

```python
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
```

# BASIC PLOTLAR 

```python
plt.plot(flights["year"], flights["passengers"])
plt.show()
```

```python
plt.step(flights["year"], flights["passengers"])
plt.show()
```

```python
plt.fill_between(flights["year"], flights["passengers"])
plt.show()
```

```python
plt.hist(titanic["age"].dropna(), bins=30)
plt.show()
```

```python
sns.histplot(titanic["fare"], kde=True)
plt.show()
```

```python
sns.ecdfplot(titanic["fare"])
plt.show()
```

```python
sns.rugplot(titanic["age"])
plt.show()
```

# categorical

```python
sns.barplot(x="sex", y="fare", data=titanic)
plt.show()
```

```python
sns.countplot(x="class", data=titanic)
plt.show()
```

```python
sns.boxplot(x="class", y="fare", data=titanic)
plt.show()
```

```python
sns.violinplot(x="class", y="fare", data=titanic)
plt.show()
```

```python
sns.stripplot(x="class", y="fare", data=titanic, jitter=True)
plt.show()
```

```python
sns.swarmplot(x="class", y="fare", data=titanic,)
plt.show()
```

```python
sns.pointplot(x="class", y="fare", data=titanic,)
plt.show()
```

```python
sns.catplot(x="class", y="fare", kind="box", data=titanic,)
```

# relationship

```python
plt.scatter(titanic["age"], titanic["fare"])
plt.show()
```

```python
sns.regplot(x="age", y="fare", data=titanic)
plt.show()
```

```python
sns.jointplot(x="age", y="fare", data=titanic)
plt.show()
```

```python
sns.jointplot(x="age", y="fare", kind="hex", data=titanic)
plt.show()
```

```python
plt.hexbin(titanic["age"], titanic["fare"], gridsize=30)
plt.colorbar(), plt.show()
```

```python
sns.pairplot(iris, hue="species")
```

# matrix 

```python
corr = titanic.select_dtypes(include="number").corr()
sns.heatmap(corr, cmap="coolwarm") 
plt.show()
```

```python
sns.clustermap(corr)
```

```python
scatter_matrix(iris, figsize=(8,8))
plt.show()
```

```python
parallel_coordinates(iris, "species")
plt.show()
```

```python
andrews_curves(iris, "species")
plt.show()
```

# Time Serties

```python
plt.plot(flights["passengers"])
plt.show()
```

```python
rolling = flights["passengers"].rolling(12).mean()
plt.plot(rolling)
plt.show()
```

```python
lag_plot(flights["passengers"])
plt.show()
```

```python
plot_acf(flights["passengers"].dropna())
plt.show()
```

```python
plot_pacf(flights["passengers"].dropna())
plt.show()
```

# proportion / ranking

```python
titanic["survived"].value_counts().plot.pie(autopct="%1.1f%%") 
plt.show()
```

```python
titanic["sex"].value_counts().plot.pie(wedgeprops=dict(width=0.4))
plt.show()
```

```python
vals = tips.groupby("day")["total_bill"].sum()
vals.cumsum().plot() 
plt.show()
```

# hierarchical 

```python
px.treemap(titanic, path=["class","sex"], values="fare") 
```

```python


px.sunburst(titanic, path=["class","sex"], values="fare") 
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(iris.iloc[:,:4], method="ward")
dendrogram(Z) 
plt.show()
```
