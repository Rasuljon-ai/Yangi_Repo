# Feature enginering -  bu xom malumotlardan model uchun foydali informative featurelar yaratish jarayoni 

```python
import pandas as pd 
df = pd.read_csv(r"C:\Users\Rasulbek907\Desktop\Project_One\Data\Raw_Data\titanic.csv")
```

```python
def missing_val(df):
    for col in df.columns:
        if df[col].isnull().sum():
            if df[col].dtype == "object":
                df[col] = df[col].fillna(df[col].mode()[0])
            else:
                df[col] = df[col].fillna(df[col].mean())
    return df 
```

```python
df = missing_val(df)
```

# Yangi Foydali Ustunlar yaratish bosqichi

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1 
```

```python
df.head()
```

```python
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
```

```python
df.head()
```

```python
df["Title"] = df["Name"].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df.head()
```

```python
df["Title"] = df["Title"].replace(['Lady', "Countess", "Capt", "Col", "Don", "Dr", "Major", "Rev", "Sir", "Jonkheer", "Rare", "Dona"])
df['Title'] = df["Title"].replace({"Mlle": "Miss", "Ms":"Miss", "Mme":"Mrs"})

df.drop(columns=["Name"], inplace=True)
```
