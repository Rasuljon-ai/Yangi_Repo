# Feature Va Targetga ajratish 

```python
x = df.drop(columns=["Cabin"])
y = df["Cabin"]
```

# Train va Temporary split == 70% ga 30 % 

```python
from sklearn.model_selection import train_test_split
x_train, x_temp, y_train, y_temp = train_test_split (x, y, test_size=0.3, random_state=42)
```

```python
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, random_state=42)
```

# Modelni o`qitish jarayoni 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()

model.fit(x_train, y_train)

y_val_pred = model.predict(x_val)
val_score = r2_score(y_val, y_val_pred)

print("Validation r2_Score :  ", val_score)
```

```python
y_test_pred = model.predict(x_test)
test_accuracy = r2_score(y_test, y_test_pred)

print("Test r2_score: ", test_accuracy)
```

```python
y_train_pred = model.predict(x_train)

train_accuracy = r2_score(y_train, y_train_pred)

print("Train r2_score: ", train_accuracy)
```
